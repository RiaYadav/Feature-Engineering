### Adding a variable to capture NaN

```python
import pandas as pd
```

```python
df=pd.read_csv('loan.csv', usecols=['BsmtQual','FireplaceQu','GarageType','SalePrice'])
df.head()
```

```python
## Lets capture the importance of Nan values in new var for 'BsmtQual' by replacing null values with '1' and others with'0'

import numpy as np
df['BsmtQual_Var']=np.where(df['BsmtQual'].isnull(),1,0)'].isnull(),1,0)
```

```python
df.head()
```

```python
## Lets update the orignal'BsmtQual' by replacinf NaN values with most frequent data
frequent=df['BsmtQual'].mode()[0]
df['BsmtQual'].fillna(frequent,inplace=True)
```

```python
'''Let do the same for 'FireplaceQu' As we are capturing importance it is usefull technique when we have 
alot of missing values but it also increases the dimensions'''


df['FireplaceQu_Var']=np.where(df['FireplaceQu'].isnull(),1,0)
frequent=df['FireplaceQu'].mode()[0]
df['FireplaceQu'].fillna(frequent,inplace=True)
```

#### Suppose if you have more frequent categories, we just replace NAN with a new category (basically creating NaN itself as a category)

```python
df=pd.read_csv('loan.csv', usecols=['BsmtQual','FireplaceQu','GarageType','SalePrice'])
df.head()
```

```python
#Lets replace NaN with "Missing" and other with their orignal values
def impute_nan(df,variable):
    df[variable+"newvar"]=np.where(df[variable].isnull(),"Missing",df[variable])
```

```python
for feature in ['BsmtQual','FireplaceQu','GarageType']:
    impute_nan(df,feature)
```

```python
df.head()
```

```python
#Lets drop the feature with NaN as we gonna use new updated version of them
df=df.drop(['BsmtQual','FireplaceQu','GarageType'],axis=1)
```
